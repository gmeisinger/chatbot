```python
import json
```

```python
num = 1
dict = {}
with open("exampleConvos.txt") as file:
    next_line = next(file)
    while next_line:
        if next_line[:3].lower() == "cov":
            dict["Cov" + str(num)] = {}
            dict["Cov" + str(num)]["Dialogue"] = []
            next_line = next(file)
            while next_line[:3].lower() != "cov":
                if next_line != "\n":
                    if next_line[:7] == "[User]:":
                        dict["Cov" + str(num)]["Dialogue"].append({"Sender":"User", "Message":next_line[8:].strip('\n'), "Intent":"","Entities":[{"Start":0,"End":0,"Value":"","Entity":""}]})
                    else:
                        dict["Cov" + str(num)]["Dialogue"].append({"Sender":"SCITalk", "Message":next_line[11:].strip('\n'), "Intent":"","Entities":[{"Start":0,"End":0,"Value":"","Entity":""}]})
                next_line = next(file)
            num+=1
```

```python
f = open("exampleConvos.json","w")
f.write(json.dumps(dict))
f.close()
```
